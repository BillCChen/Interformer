```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 示例数据
df = pd.DataFrame(df)

# 统计相同Uniprot值的行数
uniprot_counts = df['ProteinName'].value_counts()

# 取前20和最后20个值
first_20 = uniprot_counts.head(50)
last_20 = uniprot_counts.tail(50)

# 将前20和最后20个值连接在一起
selected_data = pd.concat([first_20, last_20])

# 设置图像尺寸
plt.figure(figsize=(20, 6))

# 创建barplot
sns.barplot(x=selected_data.index, y=selected_data.values, color='blue')

# 将x轴标签设置为垂直显示
plt.xticks(rotation=90, fontsize=8)

# 设置x轴和y轴的标签
plt.xlabel('ProteinNames')
plt.ylabel('Count')

# 显示图像
plt.show()
```
