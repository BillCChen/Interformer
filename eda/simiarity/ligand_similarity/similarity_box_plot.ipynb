```python
import plotly.express as px
import pandas as pd

root = '/opt/home/revoli/data_worker/paper/benchmark/docking/similarity'
input_csv = ['kinase_test.csv', 'lsd1_project.csv', 'mPro_covalent_test.csv', 'mPro_project.csv']

all_df = []
for csv in input_csv:
  df = pd.read_csv(f'{root}/{csv}')
  df['Test Set'] = csv[:-4]
  all_df.append(df)
all_df = pd.concat(all_df)

kinase_test = all_df[all_df['Test Set'] == 'kinase_test']

print(f"Kinase Num of outlier:{(kinase_test['simiarity'] > 0.70).sum()}")

colors = {}
for i, csv in enumerate(input_csv):
   colors[csv[:-4]] = px.colors.qualitative.Set1[i]
# print(colors)
fig = px.box(all_df, y="simiarity", x='Test Set', color='Test Set', color_discrete_map=colors)

fig.update_layout(
  plot_bgcolor='white',
  width=1100, height=500, font={'size': 15},
  title='The ligand similarity of affinity benchmark compare with training set',
)
fig.update_xaxes(
  title='Affinity Testset'
)
fig.update_yaxes(
  mirror=True,
  ticks='outside',
  showline=True,
  linecolor='black',
  gridcolor='lightgrey',
  title='Max Ligand Similarity',
)
fig.show()
```
